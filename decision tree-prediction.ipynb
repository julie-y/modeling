```python
def fillna_n(n):
    avg_n = df[n].astype('float32').mean(axis=0)
    return df[n].replace(np.nan,avg_n,inplace=True)


fillna_n('CTnT')
fillna_n('TC')
fillna_n('TG')
fillna_n('CRP')
fillna_n('D-dimer')
fillna_n('CK-MB')
fillna_n('NT-proBNP')

cdf = df[['Sex','Age','TC','TG','CRP','CTnT','D-dimer','CK-MB', 'NT-proBNP','Smoke','Hypertention']]
cdf.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

X = cdf.values

X
```

```python
#convert categorical variables to numerical values, with pandas.get_dummies()
from sklearn import preprocessing
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1]) 

le_TC = preprocessing.LabelEncoder()
le_BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])

le_Chol = preprocessing.LabelEncoder()
le_Chol.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3]) 
X[0:5]
```

```python
y = df["group_type"].apply(str)
y[0:5]
```

```python
#import the train/test split method
from sklearn.model_selection import train_test_split

X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)

#We will first create an instance of the DecisionTreeClassifier called drugTree.
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 7)
# it shows the default parameters

#fit the data
drugTree.fit(X_trainset,y_trainset)

#make predictions
predTree = drugTree.predict(X_testset)
print (predTree [0:5])
print (y_testset [0:5])

#Evaluate the accuracy of the prediction
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset,predTree,))
```
